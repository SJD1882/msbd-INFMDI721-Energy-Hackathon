# Electric price from web


Electricity prices for households, March 2019 (kWh, U.S. Dollar)

```python
import requests
from bs4 import BeautifulSoup
import re
import json

url = "https://www.globalpetrolprices.com/electricity_prices/"


def get_soup_from_url(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    return soup


def get_name_from_soup(soup):
    name_a_lists = soup.select('a.graph_outside_link')
    name = []
    for name_a_iterm in name_a_lists:
        name.append(name_a_iterm.text)
    return name


def get_price_from_soup(soup):
    prices = soup.find(id='commonPricesGraph')
    srcs = prices.attrs['src']

    prices = re.findall(r"data=(.+?)&tit", srcs)[0].split(',')

    prices = list(map(float, prices))
    return prices


def get_elec_data_from_url(url):
    soup = get_soup_from_url(url)
    name = get_name_from_soup(soup)
    prices = get_price_from_soup(soup)
    df_elec = pd.DataFrame(list(zip(name, prices)),
                           columns=['country', 'price'])
    return df_elec



def df_clean(df_elec):
    df_elec.loc[df_elec['country'] == 'Trinidad & Tobago',
                'country'] = 'Trinidad and Tobago'
    df_elec.loc[df_elec['country'] ==
                'Northern Macedonia', 'country'] = 'Macedonia'
    df_elec.loc[df_elec['country'] == 'Bosnia & Herz.',
                'country'] = 'Bosnia and Herzegovina'
    df_elec.loc[df_elec['country'] == 'South Korea',
                'country'] = 'Republic of Korea'

    return df_elec


def get_code(countryname):
    url = f'https://restcountries.eu/rest/v2/name/{countryname}'

    resp = requests.get(url)
    if resp.status_code != 200:
        return '404'
    resp_contents = resp.json()
    if len(resp_contents) > 1:
        for resp_content in resp_contents:
            if resp_content['name'].startswith(countryname):
                return resp_content['alpha3Code']
    return resp_contents[0]['alpha3Code']


df_elec = get_elec_data_from_url(url)

plt.barh(df_elec['country'][:21], df_elec['price'][:21])

df_elec = df_clean(df_elec)

df_elec['code'] = df_elec['country'].map(get_code)

df_elec
```
